```python
# libraries
import numpy as np
import pandas as pd

df = pd.read_csv('../data/edattainxtry_5plus.csv')
```

```python
df.head(5)
```

```python
i = df.columns[0:2]

j = (df.columns.str.startswith('aAll') + 
df.columns.str.startswith('aUrban') + 
df.columns.str.startswith('aRural')+ 
df.columns.str.startswith('aMale')+ 
df.columns.str.startswith('aFema'))

k = list(i) + list(df.columns[j])

values = df.loc[:, k]
values
```

```python

data_mod1 = pd.melt(values, 
                    id_vars=values.columns[~(values.columns.str.startswith("aMale") + values.columns.str.startswith("aFemale"))], 
                    value_vars=values.columns[values.columns.str.startswith("aMale") + values.columns.str.startswith("aFemale")],
                    var_name='sex')

data_mod2 = pd.melt(data_mod1, 
                    id_vars=data_mod1.columns[~(data_mod1.columns.str.startswith("aUrban") + data_mod1.columns.str.startswith("aRural"))], 
                    value_vars=data_mod1.columns[data_mod1.columns.str.startswith("aUrban") + data_mod1.columns.str.startswith("aRural")],
                    var_name='location', 
                    value_name='prop').reset_index(drop=True)

data_mod3 = pd.melt(data_mod2, 
                    id_vars=data_mod2.columns[~(data_mod2.columns.str.startswith("aA"))], 
                    value_vars=data_mod2.columns[data_mod2.columns.str.startswith("aA")],
                    var_name='all', 
                    value_name='Prop regardless of gender/location')
```

```python
data_mod3
```

```python
data_mod4 = data_mod3.drop(columns='value')
data_mod4[["sex", "level"]] = data_mod4.sex.str.split("_", expand=True)
data_mod4['sex'] = data_mod4['sex'].replace(['aMale', 'aFemale'], ['Male', 'Female'])
urban = ['aUrban_1','aUrban_2','aUrban_3','aUrban_4','aUrban_5','aUrban_6','aUrban_7','aUrban_8','aUrban_9']
data_mod4['location'] = data_mod4['location'].replace(urban, 'urban')
rural = ['aRural_1','aRural_2','aRural_3','aRural_4','aRural_5','aRural_6','aRural_7','aRural_8', 'aRural_9']
data_mod4['location'] = data_mod4['location'].replace(rural, 'rural')
```

```python
data_mod5 = data_mod4.drop(columns='all')
data_mod5
```

```python
data_tidy = data_mod5[['country', 'year', 'level', 'sex', 'location', 'prop', 'Prop regardless of gender/location']]
data_tidy
```

```python
data_tidy.to_csv('../data/tidy_data.csv')
```
