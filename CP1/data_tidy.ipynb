```python
# libraries
import numpy as np
import pandas as pd

df = pd.read_csv('../data/edattainxtry_5plus.csv')
```

```python
df
```

```python
i = ['country', 'year', 'gdppc2015', 'cpi2010']

j = (df.columns.str.startswith('aMalUrb') + 
df.columns.str.startswith('aMalRur')+ 
df.columns.str.startswith('aFemUrb')+ 
df.columns.str.startswith('aFemRur'))

k = list(i) + list(df.columns[j])

values = df.loc[:, k].reset_index(drop=True)
values
```

```python

data_mod1 = pd.melt(values, 
                    id_vars=values.columns[~(values.columns.str.startswith("aMal") + values.columns.str.startswith("aFem"))], 
                    value_vars=values.columns[values.columns.str.startswith("aMal") + values.columns.str.startswith("aFem")],
                    var_name='Sex_Location_Level',
                    value_name= 'prop')
```

```python
data_mod1
```

```python
data_mod2 = data_mod1.replace(['aMalUrb_1', 'aMalUrb_2', 'aMalUrb_3', 'aMalUrb_4', 'aMalUrb_5', 'aMalUrb_6', 'aMalUrb_7', 'aMalUrb_8', 'aMalUrb_9'], 
                            ['Male_Urban_1', 'Male_Urban_2', 'Male_Urban_3', 'Male_Urban_4', 'Male_Urban_5', 'Male_Urban_6','Male_Urban_7','Male_Urban_8','Male_Urban_9'])
data_mod3 = data_mod2.replace(['aMalRur_1', 'aMalRur_2', 'aMalRur_3', 'aMalRur_4', 'aMalRur_5', 'aMalRur_6', 'aMalRur_7', 'aMalRur_8', 'aMalRur_9'], 
                            ['Male_Rural_1', 'Male_Rural_2', 'Male_Rural_3', 'Male_Rural_4', 'Male_Rural_5', 'Male_Rural_6','Male_Rural_7','Male_Rural_8','Male_Rural_9'])
data_mod4 = data_mod3.replace(['aFemUrb_1', 'aFemUrb_2', 'aFemUrb_3', 'aFemUrb_4', 'aFemUrb_5', 'aFemUrb_6', 'aFemUrb_7', 'aFemUrb_8', 'aFemUrb_9'], 
                            ['Female_Urban_1', 'Female_Urban_2', 'Female_Urban_3', 'Female_Urban_4', 'Female_Urban_5', 'Female_Urban_6','Female_Urban_7','Female_Urban_8','Female_Urban_9'])
data_mod5 = data_mod4.replace(['aFemRur_1', 'aFemRur_2', 'aFemRur_3', 'aFemRur_4', 'aFemRur_5', 'aFemRur_6', 'aFemRur_7', 'aFemRur_8', 'aFemRur_9'], 
                            ['Female_Rural_1', 'Female_Rural_2', 'Female_Rural_3', 'Female_Rural_4', 'Female_Rural_5', 'Female_Rural_6','Female_Rural_7','Female_Rural_8','Female_Rural_9'])
```

```python
data_mod5[['sex','location', 'level']] = data_mod5.Sex_Location_Level.str.split("_",expand=True)
data_mod6 = data_mod5.drop(columns='Sex_Location_Level')
```

```python
data_mod6.head()
```

```python
data_tidy = data_mod6[['country', 'year', 'gdppc2015', 'cpi2010', 'location', 'sex', 'level', 'prop']]
data_tidy
```

```python
data_tidy.to_csv('../data/tidy_data.csv')
```
